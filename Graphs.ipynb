# General

Intonation contour

```python
from pandas.core.base import value_counts
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import precision_score, recall_score, f1_score


df = pd.read_csv('/content/drive/MyDrive/diploma/dfs/general experiments/df_cs_predicted.csv')

# Create a cross-tabulation matrix
confusion_matrix = pd.crosstab(df['labels'], df['predicted_labels'], rownames=['Actual'], colnames=['Predicted'])
print(value_counts(df['labels']))
# Plot the confusion matrix using seaborn
plt.figure(figsize=(10, 7))
sns.heatmap(confusion_matrix, annot=True, cmap='Blues', fmt='d')
plt.title('Confusion Matrix General (Intonation contour approach)')
plt.show()

# Filter the DataFrame to include only rows where 'labels' and 'predicted labels' are not the same
not_same_pairs = df[df['labels'] != df['predicted_labels']][['labels', 'predicted_labels']]

# Combine 'labels' and 'predicted labels' columns as pairs
not_same_pairs['pair'] = not_same_pairs['labels'] + ' - ' + not_same_pairs['predicted_labels']

# Obtain value counts of the pairs
pair_counts = not_same_pairs['pair'].value_counts()

print("Value counts of pairs where 'labels' and 'predicted_labels' are not the same:")
print(pair_counts)

labels = df['labels']
predicted_labels = df['predicted_labels']

# Calculate precision
precision = precision_score(labels, predicted_labels, average='weighted')

# Calculate recall
recall = recall_score(labels, predicted_labels, average='weighted')

# Calculate F1 score
f1 = f1_score(labels, predicted_labels, average='weighted')


# Compare the values in the 'labels' and 'predicted_labels' columns element-wise
matches = labels == predicted_labels

# Count the number of matches
num_matches = matches.sum()

# Calculate the total number of samples
total_samples = len(labels)

# Calculate the accuracy
accuracy = num_matches / total_samples

print(' Accuracy:',accuracy,'\n','F1 score:',f1,'\n','Recall score:', recall,'\n','Precision score', precision)
```
